# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF -using the equation- of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

### True/Analytical PDF


## using `numpy`

```python
#type your code here:
n_samples = 10000
X = np.random.normal(0, 1, n_samples)

sample_mean = np.mean(X)
sample_variance = np.var(X)

plt.hist(X, bins=50, density=True, alpha=0.6, color='g')

def analytical_pdf(x, mu=0, sigma=1):
    return (1/(np.sqrt(2*np.pi*sigma**2))) * np.exp(-(x-mu)**2 / (2*sigma**2))

x_values = np.linspace(-4, 4, 1000)
y_values = analytical_pdf(x_values)
plt.plot(x_values, y_values)
plt.grid(True)
plt.show()

```

## using `scipy`

```python
#type your code here:
n_samples = 10000
X = np.random.normal(0, 1, n_samples)

sample_mean = np.mean(X)
sample_variance = np.var(X)


plt.hist(X, bins=50, density=True, alpha=0.6, color='g')
x_values = np.linspace(-4, 4, 1000)
y_values = stats.norm.pdf(x_values, loc=0, scale=1) 
plt.plot(x_values, y_values)

plt.grid(True)

plt.show()
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
#type your code here:

n_samples = 10000
X = np.random.normal(0, 1, n_samples)

sample_mean = np.mean(X)
sample_variance = np.var(X)

x_values = np.linspace(-4, 4, 1000)

plt.hist(X, bins=50, density=True, alpha=0.6, color='g')
plt.plot(x_values, stats.norm.pdf(x_values, loc=0, scale=1), 'b')
plt.plot(x_values, stats.norm.cdf(x_values, loc=0, scale=1), 'r')

plt.grid(True)
plt.show()
```

C) Repeat (a) for an exponential random variable with $\lambda = 1$ using [numpy.random.exponential](https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html), and [scipy.stats.expon](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon)

```python
#type your code here:

n_samples = 10000
Y = np.random.exponential(1, n_samples) 
sample_mean = np.mean(Y)
sample_variance = np.var(Y)

plt.hist(Y, bins=50, density=True, alpha=0.6, color='g')


def analytical_pdf(x, lambd=1):
    return lambd * np.exp(-lambd * x) * (x >= 0)

x_values = np.linspace(0, 10, 1000)
y_values = analytical_pdf(x_values)
plt.plot(x_values, y_values)

plt.grid(True)
plt.show()
```

```python
#type your code here:
n_samples = 10000
X = np.random.exponential(1, n_samples) 

sample_mean = np.mean(X)
sample_variance = np.var(X)


x_values = np.linspace(0, 15, 1000)

plt.hist(X, bins=50, density=True, alpha=0.6, color='g')
plt.plot(x_values, stats.expon.pdf(x_values, scale=1), 'b', label='PDF')
plt.plot(x_values, stats.expon.cdf(x_values, scale=1), 'r', label='CDF')

plt.grid(True)
plt.legend()
plt.show()
```

# Problem 3: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.


```python
n_samples = 10000  

sample_sizes = [2, 50]  
means_2 = [np.mean(np.random.uniform(0, 1, 2)) for _ in range(n_samples)]
means_50 = [np.mean(np.random.uniform(0, 1, 50)) for _ in range(n_samples)]

plt.figure(figsize=(10, 5))
plt.hist(means_2, bins=30, alpha=0.6,  color='red', density=True)
plt.hist(means_50, bins=30, alpha=0.6, color='blue', density=True)

plt.title("Central Limit Theorem ")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
